# Abstract

In this study, we sought to determine the correlation between the average temperature of a state and the number of Division I NCAA basketball players it produced between the 2018-2019 and 2023-2024 seasons. Additionally, we investigated how these temperatures relate to the players' performance, measured by their average points, rebounds, and assists per game during the same period. This inquiry is driven by the hypothesis that warmer climates might facilitate greater participation in outdoor sports, potentially leading to a higher number of talented basketball players.

To explore this hypothesis, we utilized state-level data on average temperatures and the number of Division I basketball players. We employed Ordinary Least Squares (OLS) regression analysis to examine the potential correlations between these variables. Our analysis revealed a statistically significant positive relationship between a state's average temperature and the number of Division I basketball players it produced, with warmer states generally producing more players. Specifically, for each degree increase in average temperature, there was an approximate increase of 5.82 players. However, the R-squared value of 0.264 indicates that temperature explains only 26.4% of the variance in player counts, suggesting other factors also play a significant role.

When analyzing the correlation between average temperature and player performance metrics (points, rebounds, and assists per game), the results were less conclusive. The regression analyses showed weak and statistically insignificant relationships between temperature and these performance metrics. For instance, the relationship between average temperature and points per game had an R-squared value of only 0.055, indicating that temperature accounted for a mere 5.5% of the variation in points per game. Similar weak relationships were observed for rebounds and assists per game.

In conclusion, while our findings suggest a positive correlation between warmer climates and the number of Division I basketball players produced, the impact of temperature on player performance metrics appears minimal and statistically insignificant. These results highlight the complexity of factors influencing the development of basketball talent and suggest that further research incorporating additional variables, such as socioeconomic factors and access to training facilities, is necessary to gain a comprehensive understanding of what drives basketball success at the collegiate level.

```python
# Remove data points not from 2018 to 2023
weather_df = weather_df[weather_df["code"].str[6:].isin(["2018", "2019", "2020", "2021", "2022", "2023"])]
```

```python
# Convert state codes into their state name strings
weather_df["state"] = weather_df["state"].map(state_codes)
```

# Discussion and Conclusion
In this project, we sought to uncover whether the average temperature of a state affects the number of college level basketball players that the state produces. Moreover, we also aimed to explore the relationship between weather and the players’ performance. Before we take a look at the results of our exploration, we first need to establish a significance level for the analysis that we carried out. We decided on 0.10 since it reduces the likelihood of making a type two error, which may cause us to wrongly conclude that there is no correlation between two variables when there actually may be.

Our main question aims to discover whether each states’ average temperatures have an impact on the total number of division one college basketball players they produce. We collected data spanning 6 years from the 2018-19 season to the 2023-24 season. This totaled to 28885 players from the United States. The linear regression line has a coefficient of around 5.82, which means that with every one degree Fahrenheit increase in the average temperature, the state produces around 6 more players. A p-value of 0.000 ascertains that we reject the null hypothesis that there is no correlation between temperature and the number of players. This finding echoes the conclusion in Chan and Ryan’s paper: physical activity levels are generally higher on warmer days and lower during inclement weather. This would make sense as higher levels of physical activity means more time spent playing basketball and developing skills, thus more likely to produce division one caliber basketball players.

The sub-question investigates the relationship between temperature and players’ performance. We considered the three main statistical categories in basketball – points per game, assists per game, and rebounds per game. We used data on the NCAA official site for each season’s stat leaders and recorded each player’s birth state. To properly explore the relationship between these statistics and the state’s temperature, we calculated the average of the statistical category for each state over the 6 seasons. Plotting average temperatures of each state vs. the statistical categories revealed coefficients of -0.038, -0.011, and 0.009 for points, assists, and rebounds, respectively. The p-values for each respective category are 0.120, 0.202, and 0.282. These values are all too high for us to be able to reject the null hypothesis, so we are unable to conclude that temperatures have an impact on the performance of college basketball players. Even if the p-values were to be below the significance level, the coefficients of the regression lines are fairly close to 0. In a way, this attests to how well NCAA upholds their standards for player selection.

This includes taking into account a state’s population, a state’s attractiveness for aspiring basketball players, accessibility to indoor facilities, and much more in order to account for the various factors that may impact an individual’s likelihood of becoming a college-level player. Although we were unable to conclude a correlation for how player performance is affected by temperature, our study only approached this question from one direction, and there are many more angles to approach this exploration. These may include studying the max value of the performance statistical categories rather than the average, taking into account how many players from those states are actively playing in college basketball, or by considering the age and experience of individual players.
